# 3. Branch PR Workflow

Using multiple branches and a PR approach to merge in work enables multiple users to work on different features of the same project at the same time and also ensures cleaner and more efficient code. Branches help users work independently on their features without affecting any other users work. The PR reviews ensure that the code that is merged in is free of any bad code (as it goes through multiple sets of eyes instead of just the person working on that specific feature)
